```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
## Line Plot
```

```python
var1 = [i for i in range(1,8)]
var2 = [i**3 for i in range(1,15,2)]
```

```python
plt.plot(var1,var2)
plt.show()
```

```python

x=pd.DataFrame({'var1':var1,'var2':var2})
sns.lineplot(x='var1',y='var2',data=x)
```

```python
data = pd.read_csv('/Users/nanda/Downloads/car data.csv')
data
```

```python
sns.lineplot(x='Year',y='Selling_Price',data=data,hue=data.Fuel_Type,size=20,style='Fuel_Type',palette='Accent'
             ,markers=['o','^'],dashes=False,legend='brief',    )
```
