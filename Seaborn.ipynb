```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
## Line Plot
```

```python
var1 = [i for i in range(1,8)]
var2 = [i**3 for i in range(1,15,2)]
```

```python
plt.plot(var1,var2)
plt.show()
```

```python

x=pd.DataFrame({'var1':var1,'var2':var2})
sns.lineplot(x='var1',y='var2',data=x)
```

```python
ar = np.array([['a0','a1','a2','a3','a4'],['b0','b1','b2','b3','b4']])
ar
```

```python
y={'fontsize':20,'color':'r'}
v = sns.heatmap(var,vmin=0,vmax=10,cmap='Accent',annot=ar,fmt='s',annot_kws=y,linewidths=10,linecolor='b',cbar=False,xticklabels=False,yticklabels=False)
v.set(xlabel='python')
sns.set(font_scale=1)
```
